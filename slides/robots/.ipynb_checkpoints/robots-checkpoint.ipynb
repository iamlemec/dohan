```python
import os
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout, to_agraph
import pygraphviz as pgv
import nxpd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
slide_dir = '/home/doug/Dropbox/Discussant/Berg_Buffie_Zanna_CRNYU/slides'
slide_path = lambda fn: os.path.join(slide_dir, fn)
```

```python
# Model 1
G = nx.DiGraph()
G.add_edge('Capital (K)', 'Output (Q)', label='sigma1')
G.add_edge('Composite (V)', 'Output (Q)', label='sigma1')
G.add_edge('Labor (L)', 'Composite (V)', label='sigma2')
G.add_edge('Robots (bZ)', 'Composite (V)', label='sigma2')
G.graph['graph']={'rankdir':'TD'}
G.graph['node']={'shape':'box'}
G.graph['edges']={'arrowsize':'4.0'}
nxpd.draw(G, filename=slide_path('model_1.svg'), show=False)
nxpd.draw(G, show='ipynb')
```
